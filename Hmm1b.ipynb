```python
import nltk
from nltk.corpus import brown
from nltk.tag import hmm
import random
```

```python
nltk.download('punkt')
```

```python
nltk.download('brown')
```

```python
tagged_sentence = brown.tagged_sents()
```

```python
sentences = tagged_sentence[:1000]
```

```python
print(sentences[0])
```

```python
train = hmm.HiddenMarkovModelTrainer()
hmm_model = train.train_supervised(sentences)
```

```python
def generate_sentence(hmm_model,length):
  words=[]
  state = random.choice(list(hmm_model._states))
  for i in range(length):
    word = hmm_model._outputs[state].generate()
    words.append(word)
    state=hmm_model._transitions[state].generate()
  return words
```

```python
for i in range(5):
    generate = generate_sentence(hmm_model,10)
    print(" ".join(generate))
    print()
```
