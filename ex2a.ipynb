```python
import nltk
from nltk.corpus import brown
from gensim.models import Word2Vec
```

```python
nltk.download('brown')
```

```python
sentences = brown.sents()[:5000]

# convert to lowercase
sentences = [[word.lower() for word in sent] for sent in sentences]

print("Total sentences:", len(sentences))
```

```python
query_words = ["money", "government", "student", "computer"]
```

```python
def train_model(vector_size, window):

    model = Word2Vec(
        sentences,
        vector_size=vector_size,
        window=window,
        sg=1,            # Skip-gram
        min_count=2,
        workers=4
    )

    return model
```

```python
settings = [
    (50, 3),
    (50, 5),
    (100, 3),
    (100, 5)
]

models = {}

for vec, win in settings:
    print(f"\nTraining model: vector_size={vec}, window={win}")
    models[(vec, win)] = train_model(vec, win)
```

```python
model = models[(50,3)]

print("\nEmbedding vector for word 'money':")
print(model.wv['money'][:10])
```

```python
for params, model in models.items():

    vec, win = params
    print("\n==============================")
    print(f"Vector Size: {vec} | Window: {win}")
    print("==============================")

    for word in query_words:

        if word in model.wv:
            print(f"\nTop 8 similar words to '{word}':")

            similar_words = model.wv.most_similar(word, topn=8)

            for w, score in similar_words:
                print(w, round(score,3))

        else:
            print(f"{word} not in vocabulary")
```

```python
import nltk
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt_tab')
tokenized_sentences = [word_tokenize(sent.lower()) for sent in sentences]

print("Tokenized Sentences:")
print(tokenized_sentences)
```

```python
model = Word2Vec(
tokenized_sentences,
vector_size=50,
window=3,
min_count=1,
sg=1
)
```

```python
print("\nVocabulary:")
print(list(model.wv.index_to_key))
```

```python
print("\nEmbedding for word 'student':")
print(model.wv['student'])
```

```python
query_words = ["student","computer","money"]
```

```python
for word in query_words:
    print(f"\nTop 5 similar words to '{word}':")
    similar_words = model.wv.most_similar(word, topn=5)

    for w, score in similar_words:
        print(w, round(score,3))
```
