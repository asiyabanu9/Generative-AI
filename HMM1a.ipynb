```python
import random

def generate_random(hmm_model, length=10):
    sentence = []

    # initial hidden state (POS tag)
    state = random.choice(list(hmm_model._states))

    for _ in range(length):
        # emission: state -> word
        word = hmm_model._outputs[state].generate()
        sentence.append(word)

        # transition: state -> next state
        state = hmm_model._transitions[state].generate()

    return sentence
```

```python
# @title
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk import pos_tag
from nltk.tag import hmm
import nltk, random

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

text = """
I love learning new things
Machine learning is interesting
Data science helps students
I enjoy learning models
Artificial intelligence is powerful
"""

sentences = sent_tokenize(text)

tagged_sentences = []
for sent in sentences:
    words = word_tokenize(sent)
    tagged_sentences.append(pos_tag(words))

print("Tagged sentences:", tagged_sentences)
```

```python
# @title
trainer = hmm.HiddenMarkovModelTrainer()
hmm_model = trainer.train_supervised(tagged_sentences)

print("States learned:", hmm_model._states)
```

```python
# @title
def generate_sentence(model, length=10):
    sentence = []
    state = random.choice(list(model._states))

    for _ in range(length):
        word = model._outputs[state].generate()
        sentence.append(word)
        state = model._transitions[state].generate()

    return sentence

generated = generate_sentence(hmm_model, 10)
print("Generated sentence:")
print(" ".join(generated))
```
